```python
import pandas as pd
```

```python
from pycaret.classification import setup, compare_models, evaluate_model, save_model, load_model, predict_model
```

```python
churn_data = pd.read_csv('/content/updated_churn_data (1).csv')
```

```python
clf = setup(data=churn_data, target='Churn')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
save_model(best_model, 'best_churn_model')
```

```python
def predict_churn_probability_whole_dataset(data):
  model = load_model('best_churn_model')

  predictions = predict_model(model, data=churn_data)

  return predictions
```

```python
new_data=pd.read_csv('/content/new_churn_data.csv')
```

```python
predictions = predict_churn_probability_whole_dataset(new_data)
```
